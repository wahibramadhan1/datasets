```python
import pandas as pd #load libraries
from sklearn.tree import DecisionTreeClassifier #import Decision Tree Classifier
from sklearn.model_selection import train_test_split #import train_test_split function
from sklearn import metrics #import scikit-learn metrics module for accuracy calculation
```

```python
pd.__version__ #melihat versi pandas saat ini
```

```python
pd.read_csv("heart.csv") #membaca file heart dalam folder datasets
```

```python
sehat=pd.read_csv("heart.csv")  #membuat variabel sehat dari dataframe heart
sehat
```

```python
sehat.head(5) #mengecek lima baris paling atas
```

```python
sehat.info() #mengecek nama kolom dalam varibel sehat
```

```python
keter = ['age','sex','cp','trestbps','chol','fbs','restecg','thalach','exang','oldpeak','slope','ca','thal','target']
for lol in keter: #coerce for missing values
    sehat[lol]=pd.to_numeric(sehat[lol],errors='coerce')
```

```python
sehat.dropna(inplace=True)
```

```python
lol_tawa = ['age','sex','cp','trestbps','chol','fbs','restecg','thalach','exang','oldpeak','slope','ca','thal'] #split dataset in features and target variable
X = sehat[lol_tawa] #features
y = sehat.target #variable target
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3,random_state=1) #split dataset into training set and test set and also 70% training and 30% test
```

```python
gan=DecisionTreeClassifier() #create decision tree calssifer object
gan=gan.fit(X_train,y_train) #train decision tree classifier
y_pred = gan.predict(X_test) #predict the response for test dataset
```

```python
print("Accuracy:",metrics.accuracy_score(y_test,y_pred))
```
